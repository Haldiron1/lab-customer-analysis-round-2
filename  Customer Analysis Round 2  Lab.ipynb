```python
df4 = df.dropna(subset=['vehicle type'])
```

```python
df4
```

```python
df4.isna().sum()
```

```python
df4.columns = df4.columns.str.replace(' ', '_')
```

```python
df4 = df4.rename(columns={'employmentstatus':'employment_status'})
```

```python
df4.columns
```

```python
df4.isna().sum()
```

```python
median1 = df4["months_since_last_claim"].median()
median2 = df4["number_of_open_complaints"].median()
df4["months_since_last_claim"] = df4["months_since_last_claim"].fillna(median1)
df4["number_of_open_complaints"] = df4["number_of_open_complaints"].fillna(median2)
```

```python
df4.isna().sum()
```

```python
df4['response'] = df4['response'].fillna('No Response')
```

```python
df4['state'] = df4['state'].fillna('Other')
```

```python
df4.isna().sum()
```

```python
df4['vehicle_class'] = df4['vehicle_class'].fillna('Unknown')
df4['vehicle_size'] = df4['vehicle_size'].fillna('Unknown')
```

```python
df4.isna().sum()
```
